```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b

def y(x, a, b): 
    return a*x + b

x1,_ = curve_fit(y, x, y1)
x2,_ = curve_fit(y, x, y2)

# Red line is line that fit to each points

plt.figure(0)
plt.plot(x, y1, 'b.')
plt.plot(x, y(x, *x1), 'r.')
plt.show()

plt.figure(1)
plt.plot(x, y2, 'k--')
plt.plot(x, y(x, *x2), 'r')
plt.show()


# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```
